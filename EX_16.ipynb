```python
# üì¶ Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import dendrogram, linkage

# üì• Load the Iris dataset
iris = pd.read_csv("/content/IRIS.csv")

# üîç Separate features and labels
x = iris.iloc[:, :-1].values  # Features
y = iris.iloc[:, -1].values   # Labels (not used in clustering)

# üîÄ Optional: Split data (not used in clustering but included for completeness)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# üåø Apply Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=3)
y_pred = hc.fit_predict(x)

# üìä Scatter plot of clusters (2D using first two features)
plt.figure(figsize=(8, 5))
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s=50, c='blue', label='Cluster 3')
plt.title("Agglomerative Clustering on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# üå≤ Dendrogram for hierarchical structure
plt.figure(figsize=(10, 6))
plt.title("Dendrogram (Hierarchical Clustering)")
dendrogram(linkage(x, method='ward'))
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
```
