```python
# üì¶ Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

# üì• Load the Iris dataset
iris = pd.read_csv("/content/IRIS.csv")

# üîç Separate features and labels
x = iris.iloc[:, :-1].values  # Features (first 4 columns)
y = iris.iloc[:, -1].values   # Labels (last column, unused in clustering)

# üîÄ Optional: Split data (not used in clustering but included for completeness)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# ü§ñ Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(x)

# üìå Print cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)

# üìä Create 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# üå∏ Plot clusters
ax.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], x[y_pred == 0, 2], c='red', label='Cluster 1')
ax.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], x[y_pred == 1, 2], c='green', label='Cluster 2')
ax.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], x[y_pred == 2, 2], c='blue', label='Cluster 3')

# ‚≠ê Plot centroids
ax.scatter(kmeans.cluster_centers_[:, 0],
           kmeans.cluster_centers_[:, 1],
           kmeans.cluster_centers_[:, 2],
           s=500, c='yellow', label='Centroids')

# üè∑ Final touches
plt.legend()
plt.title("KMeans Clustering on Iris Dataset (3D Visualization)")
plt.show()
```
